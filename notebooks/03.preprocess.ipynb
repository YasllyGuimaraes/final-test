### Importando bibliotecas:

```python
import pandas as pd
from nltk.corpus import stopwords
stopwords = stopwords.words('english')
import pickle
```

### Carregando datasets:

```python
# Carregando data frame de treino com a coluna acrescentada:

with open("train_df_mod.pkl", "rb") as files:
    train_df = pickle.load(files)
train_df.head(5)
```

```python
# Carregando dataframe de teste com a coluna acrescentada:

with open("test_df_mod.pkl", "rb") as files:
    test_df = pickle.load(files)
test_df.head(5)
```

### Aplicando mais pr√©-processamentos e criando colunas:

```python
# Importando arquivo limpando dados:

from configs.limpando_dados import Limpeza
```

Dados de treino:

```python
# pre-processado + lemmatizacao:
train_df['pre-processado-lemma'] = train_df['pre-processado'].apply(
    lambda x: Limpeza.lemmatizacao(x))

# pre-processado + stemming:
train_df['pre-processado-stem'] = train_df['pre-processado'].apply(
    lambda x: Limpeza.stemming(x))

# pre-processado + lemmatizacao + stopwords:
train_df['pre-processado-sw-lemma'] = train_df['pre-processado'].apply(
    lambda x: Limpeza.lemmatizacao(x))
train_df['pre-processado-sw-lemma'] = train_df['pre-processado-sw-lemma'].apply(
    lambda x: Limpeza.remover_stopwords(x, stopwords))

# pre-processado + lemmatizacao + stopwords:
train_df['pre-processado-sw-stem'] = train_df['pre-processado'].apply(
    lambda x: Limpeza.stemming(x))
train_df['pre-processado-sw-stem'] = train_df['pre-processado-sw-stem'].apply(
    lambda x: Limpeza.remover_stopwords(x, stopwords))

train_df.head(5)
```

Dados de teste:

```python
# pre-processado + lemmatizacao:
test_df['pre-processado-lemma'] = test_df['pre-processado'].apply(
    lambda x: Limpeza.lemmatizacao(x))

# pre-processado + stemming:
test_df['pre-processado-stem'] = test_df['pre-processado'].apply(
    lambda x: Limpeza.stemming(x))

# pre-processado + lemmatizacao + stopwords:
test_df['pre-processado-sw-lemma'] = test_df['pre-processado'].apply(
    lambda x: Limpeza.lemmatizacao(x))
test_df['pre-processado-sw-lemma'] = test_df['pre-processado-sw-lemma'].apply(
    lambda x: Limpeza.remover_stopwords(x, stopwords))

# pre-processado + lemmatizacao + stopwords:
test_df['pre-processado-sw-stem'] = test_df['pre-processado'].apply(
    lambda x: Limpeza.stemming(x))
test_df['pre-processado-sw-stem'] = test_df['pre-processado-sw-stem'].apply(
    lambda x: Limpeza.remover_stopwords(x, stopwords))

test_df.head(5)
```

### Salvando dataframe com todas as colunas:

```python
# Salvanda dataframe de treino:

with open('train_df_final.pkl', 'wb') as files:
    pickle.dump(train_df, files)
```

```python
# Salvando dataframe de teste:

with open('test_df_final.pkl', 'wb') as files:
    pickle.dump(test_df, files)
```
